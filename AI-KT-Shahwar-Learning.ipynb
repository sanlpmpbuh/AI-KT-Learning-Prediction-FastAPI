**Execution Flow**

* Python first compiles your source code (the statements in your file) into a format known
as byte code.
* Byte code is a lower-level,
platform-independent representation of your source code.
* This compilation is just Translation.
* These byte code (extn : .pyc), translated from source code (.py).
* Every time you save your script, new byte code is created.
* Every time time you exit, byte codes are discarded.


**Why this conversion?**

1. To speed up the 
executionâ€” Byte code can be run much more quickly than the original source code(lines of script).

**User-Defined Classes**

```python
class Worker:
    def __init__(self,name,pay):
        self.name = name
        self.pay = pay
        
    def lastName(self):
        return self.name.split()[-1]
    
    def giveRaise(self,percent):
        self.pay = self.pay * (1 + percent)
        return self.pay
```

```python
#Let us make OBJECTS

bob = Worker('Bob Smith', 45000)
tony = Worker('Tony Stark', 56000)
```

```python
bob.lastName()
```

```python
tony.lastName()
```

```python
bob.giveRaise(.10)
```
